## Groupby

```python
groups = df.groupby("education")
```

```python
groups
```

```python
# the same as df[df["education"] == "Masters"}]
groups.get_group("Masters")
```

Value_counts

```python
groups.value_counts().to_frame()
```

Mean age for every group

```python
groups.age.mean().sort_values().info
```

Aggregation

```python
 groups.agg({"age":"mean"})
```

```python
age_df = groups.agg({"age": ["count", "mean", "std"]})
age_df
```

```python
age_df["age"]["count"]
```

```python
# shows structure of multiIndex columns
age_df.columns
```

```python
# joins 
age_df.coumns = age_df.columns.map(' _ '.join)
```

```python
age_df
```

```python
groups.agg({"age": ["min", "max"], "hours-per-week": "median"})
```

Standard error

```python
# use lambda fuction to find a quadratic mean (root square mean)
groups.agg({
    "age": ["min", "max"],
    "hours-per-week": "median",
    "capital-gain": [("standard-error", lambda x: np.std(x) / np.sqrt(len(x)))]
})
```
