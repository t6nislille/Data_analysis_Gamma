```python
import numpy as np
from matplotlib import pyplot as plt
```

## 1. Plot creation and basic configuration

1.1 Create of a linaer plot

```python
x = np.arange(3, 9)
y = x**2
x2 = np.array([1, 2, 3])
y2 = x2 * 10 + 3
print(x)
print(y)
print(x2)
print(y2)
```

```python
plt.plot(x, y, 'o-g')
plt.plot(x2, y2, color="blue", linewidth=1, marker='o',
         markersize=3, markeredgewidth=4, alpha=0.6)
plt.grid()
```

#### 1.2 Setp

```python
lines = plt.plot(x, y, x2, y2)
plt.setp(lines, c="#bcc953", lw=3, marker="<", ms=8,
         mfc="aquamarine", ls="-.", gapcolor="b", alpha=0.5)
plt.grid()
```

## 2. Subplots

2.1 Add subplot

```python
plt.subplot(2, 2, 2) # cordinates 1 parameter= rows, 2 parameter= columns, 3 parameter= number of graph
plt.plot(np.random.rand(10))
plt.grid()
plt.subplot(2, 2, 3)
plt.plot(np.random.rand(10))
plt.grid()
plt.subplot(2, 2, 1)
plt.plot(np.arange(9))
plt.grid()
```

```python
fig, ax = plt.subplot(2, 2, sharex=True, shareY=True, width_ratios=[0.3, 0.7], height_ratios=[0.6, 0.4])

fig.set_size_inches(12, 7)
fig.set_facecolor("#b296eb")
fig.set_edgecolor("#orange")
fig.set_linewidth(5)

ax[0, 1].plot(np.sin(np.range(-np.pi, 0,1)), "x-r")
```

#### 2.3 Gridspec

```python
fig = plt.figure()
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[1:, 0])
ax1 = fig.add_subplot(gs[0, :])
ax3 = fig.add_subplot(gs[1:, 2])
```

```python
fig = plt.figure()
gs = fig.add_gridspec(4, 3)
ax1 = fig.add_subplot(gs[:, 0])
ax1 = fig.add_subplot(gs[0, 1:])
ax3 = fig.add_subplot(gs[1, 1])
ax4 = fig.add_subplot(gs[2:, 1])
ax5 = fig.add_subplot(gs[1:3, 2])
ax6 = fig.add_subplot(gs[3, 2])
```

```python
fig = plt.figure()
gs = fig.add_gridspec(4, 4)
ax1 = fig.add_subplot(gs[:2, 0])
ax1 = fig.add_subplot(gs[2:, 0])
ax3 = fig.add_subplot(gs[0, 1:3])
ax4 = fig.add_subplot(gs[1, 1])
ax5 = fig.add_subplot(gs[1, 2])
ax6 = fig.add_subplot(gs[2, 1:3])
ax7 = fig.add_subplot(gs[3, 1])
ax8 = fig.add_subplot(gs[3, 2])
ax9 = fig.add_subplot(gs[:, 3])
plt.tight_layout()
```

## 3. Grids & limits

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), "b--")
plt.xlim(-50, 250)
plt.ylim(-100, 100)
# ax.set_xlim(-50, 250)
# ax.set_ylim(-100, 100)
```

```python
fig.plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.tan(np.arange(-10, 10, 0.1)), "b--")
ax.axis((0, 200, -80, 80)) # set limits for both axis: x and y
```

### 4. Titles

```python
font1 = {"family": "comic sans ms", "color": "brown", "size": 15, "weight": "bold"}
font2 = {"family": "times new roman", "color": "#49b1de", "size": 12, "weight": "italic"}

fig, ax = plt.subplots()
ax.plot(np.cos(np.arange(-np.pi, np.pi, 0,2)), "-.o")
plt.title("THIS IS A TITLE!", loc="right", fontdict=font1)
```

## 5. Legend

```python
fig.plt.figure(figsize=(8,5))
ax = fig.add_subplot()
ax.plot(np.sin(np.linspace(-np.pi/2, 20)), "cx:", label="Sinewave")
```

## 6. Fill between

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y)
```

## 9. Histogram

```python
fig, ax = plt.subplots()

y = np.random.normal(0, 2, 1000)
ax.hist(y)
ax.grid()
```

```python
fig, ax = plt.subplots()

y = np.random.normal(0, 2, 1000)
ax.hist(y, color="lime", edgecolor="black", bins=25)
ax.set_xticks(np.arange(-7, 8))
ax.grid()
```

```python
# set bins manualy
fig, ax = plt.subplots()

y = np.random.normal(0, 2, 1000)
ax.hist(y, color="lime", edgecolor="black", bins=[-5, -1, -0.5, 0, 0.5, 1, 2, 4, 8])
ax.set_xticks(np.arange(-7, 8))
ax.grid()
```

## 10. Bar plot

```python
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot()

x = np.arange(1, 11)
y = np.random.randint(-20, 20, 10)

ax.bar(x, y)
ax.grid()
```

## 11. Pie plot

```python
fig, ax = plt.subplots(figsize=(7, 7))

l = ["BMW", "Kia", "Volvo", "Toyota", "Subaru"]
v = [21, 14, 17, 28, 30]

ax.pie(v, labels=l)
plt.show() # for PyCharm
```

## 12. Stem plot

```python
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot()

x = np.arange(-np.pi, np.pi, 0.2)
ax.stem(x, np.cos(x), ":g", markerfmt="dr", basefmt="--p", bottom=0.25)
```

## 13. Scatter plot

```python
fig, ax = plt.subplots()

x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y, s=50, c="silver", edgecolor="b", marker="h")
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.random.normal(0, 2, 50)
y = np.random.normal(0, 2, 50)
sizes = np.random.uniform(30, 300, 50)
colors = np.random.uniform(10, 300, 50)

p = ax.scatter(x, y, s=sizes, c=colors, cmap="Spectral", alpha=0.55)
ax.grid()

fig.colorbar(p)
```

## 14. Boxplot

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.random.normal((3, 5, 4), (1.25, 1, 1.5), (100, 3))
ax.boxplot(x)
ax.grid()
```
